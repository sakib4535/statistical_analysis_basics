```python
import scipy.stats as st
import matplotlib.pyplot as plt
import numpy as np
```

```python
xs = np.linspace(-5, 20, 100)
ks = np.arange(50)
```

```python
#Discrete()

pmf_binom = st.binom.pmf(ks, 50, 0.25) # 50 events and each event has 25% chances of successfully happening.
plt.bar(ks, pmf_binom, label="Binominal Distribution", alpha=0.9)

pmf_poisson = st.poisson.pmf(ks, 35)
plt.bar(ks, pmf_poisson, label="Binominal Distribution", alpha=0.9)
plt.xlabel("Events")
plt.ylabel("Possibility Rate")
plt.legend()

print(st.binom.pmf(10, 50, 0.25))
print(st.poisson.pmf(50, 30))
```

```python
#Continuous Distribution

uniform = st.uniform.pdf(xs, -4, 10)
plt.plot(xs, uniform, label="Uniform(-4,6)")

normal = st.norm.pdf(xs, 5, 2)
plt.plot(xs, normal, label="Normal(5,2)")

exponential = st.expon.pdf(xs, loc=5, scale=2)
plt.plot(xs, exponential, label="Exponential(0.5)")

student = st.t.pdf(xs, 1)
plt.plot(xs, student, label="Student t-(1)")

lognorm = st.lognorm.pdf(xs, 1)
plt.plot(xs, lognorm, label="Lognorm(1)")

skewnorm = st.skewnorm.pdf(xs, -6)
plt.plot(xs, skewnorm, label="Skewnorm")

plt.legend();
plt.show()
```

```python
plt.plot(xs, st.t.pdf(xs, 1, loc=4, scale=2), label="Inbuilt")
plt.plot(xs, st.t.pdf((xs-4) / 2, 1, loc=0, scale=2), label="manually")
```
