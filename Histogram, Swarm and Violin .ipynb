```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
```

```python
df1 = np.loadtxt("example_1.txt")
df2 = np.loadtxt("example_2.txt")
print(df1.shape, df2.shape)
```

```python
plt.hist(df1, label="Data 1")
plt.hist(df2, label="Data 2")
plt.legend()
plt.ylabel("Counts")
```

```python
counts1, bins, _ = plt.hist(df1, bins=15, label="Data 1")
plt.hist(df2, bins=bins, label="Data 2")
plt.legend()
plt.ylabel("Counts")
```

```python
bins = np.linspace(min(df1.min(), df2.min()), max(df1.max(), df2.max()), 60)
counts1, _, _ = plt.hist(df1, bins=15, label="Data 1")
plt.hist(df2, bins=bins, label="Data 2")
plt.legend()
plt.ylabel("Counts");
```

```python
bins = np.linspace(min(df1.min(), df2.min()), max(df1.max(), df2.max()), 60)
plt.hist([df1, df2], bins=bins, label="Stacked", density=True, alpha=0.5)
counts1, _, _ = plt.hist(df1, bins=15, label="Data 1", density=2, histtype="step", lw=3)
plt.hist(df2, bins=bins, label="Data 2", density=2, histtype="step", ls=":")
plt.legend()
plt.ylabel("Probability");
```

```python
bins = np.linspace(min(df1.min(), df2.min()), max(df1.max(), df2.max()), 60)
plt.hist([df1, df2], bins=bins, label="Stacked", density=True, histtype="barstacked", alpha=0.5)
counts1, _, _ = plt.hist(df1, bins=15, label="Data 1", density=2, histtype="step", lw=3)
plt.hist(df2, bins=bins, label="Data 2", density=2, histtype="step", ls=":")
plt.legend()
plt.ylabel("Probability");
```

```python
dataset = pd.DataFrame({
    "value": np.concatenate((df1, df2)),
    "type": np.concatenate((np.ones(df1.shape), np.zeros(df2.shape)))
    
})

dataset.info()
```

```python
sb.swarmplot(dataset["value"])
```

```python
sb.swarmplot(x="type", y="value", data=dataset, size=2)
```

```python
sb.boxplot(x="type", y="value", data=dataset)
sb.swarmplot(x="type", y="value", data=dataset, size=2, color="k", alpha=0.3)
```

```python
#violin plot

sb.violinplot(x="type", y="value", data=dataset);
sb.swarmplot(x="type", y="value", data=dataset, size=2, color="k", alpha=0.3);
```

```python
sb.violinplot(x="type", y="value", data=dataset, inner="quartile", bw=0.3)
```

```python
#Empirical cumulative distribution function- CDF

sd1 = np.sort(df1)
sd2 = np.sort(df2)
cdf = np.linspace(1/df1.size, 1, df1.size)

plt.plot(sd1, cdf, label="Data 1")
plt.plot(sd2, cdf, label="Data 2")
plt.hist(df1, histtype="step", density=True, alpha=0.3)
plt.hist(df2, histtype="step", density=True, alpha=0.3)
plt.legend();
plt.show()
```
