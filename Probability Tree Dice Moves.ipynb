```python
import numpy as np
import matplotlib.pyplot as plt
data = np.loadtxt("loaded_500.txt")
```

```python
unique, counts = np.unique(data, return_counts=True)
print(unique, counts)
plt.hist(data)
```

```python
num_sixes = (data == 6).sum()
num_total = data.size
```

```python
from scipy.stats import binom
n = np.arange(num_total)
prob_n = binom.pmf(n, num_total, 1/6)
plt.plot(n, prob_n,color="r",label="Prob Number")
plt.axvline(num_total/6, ls="--", lw=1, label="Mean Number")
plt.axvline(num_sixes, ls=":", color="#ff7272", label="OBS Number")
plt.xlabel(f"Num sixes {num_total}")
plt.ylabel("Probability")
plt.legend();
```

```python
d = binom(num_total, 1/6)
plt.plot(n, d.sf(n))
plt.axvline(num_sixes, c="r", ls="--")
sf = d.sf(num_sixes)
plt.axhline(sf, ls="--")
plt.xlabel("Num sixes")
plt.ylabel("SF");
print(f"Only{sf * 100:.1f}% of the time with a fair dice you'd roll this many or more sixes.")
```
